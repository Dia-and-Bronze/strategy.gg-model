```python
import pandas as pd
import numpy as np
# All In One
tier, start, end = 'GOLD', 2, 50
print("{} start".format(tier))
for m in range(start, end+1):
    print(" >> Processing : RNN_{} training start".format(m))
    try:
        data1 = pd.read_pickle("RNN Real Dataset/{0}/{1}/Data/{1}{2}.pkl".format("Original", tier, m))
        data2 = pd.read_pickle("RNN Real Dataset/{0}/{1}/Train/Data/{1}{2}.pkl".format("Additional", tier, m))
        data3 = pd.read_pickle("RNN Real Dataset/{0}/{1}/Test/Data/{1}{2}.pkl".format("Additional", tier, m))
        targetVector1 = pd.read_pickle("RNN Real Dataset/{0}/{1}/Target/{1}{2}.pkl".format("Original", tier, m))
        targetVector2 = pd.read_pickle("RNN Real Dataset/{0}/{1}/Train/Target/{1}{2}.pkl".format("Additional", tier, m))
        targetVector3 = pd.read_pickle("RNN Real Dataset/{0}/{1}/Test/Target/{1}{2}.pkl".format("Additional", tier, m))
        data = np.array(list(data1)+list(data2)+list(data3))
        targetVector = np.array(list(targetVector1)+list(targetVector2)+list(targetVector3))
    except:
        data2 = pd.read_pickle("RNN Real Dataset/{0}/{1}/Train/Data/{1}{2}.pkl".format("Additional", tier, m))
        data3 = pd.read_pickle("RNN Real Dataset/{0}/{1}/Test/Data/{1}{2}.pkl".format("Additional", tier, m))
        targetVector2 = pd.read_pickle("RNN Real Dataset/{0}/{1}/Train/Target/{1}{2}.pkl".format("Additional", tier, m))
        targetVector3 = pd.read_pickle("RNN Real Dataset/{0}/{1}/Test/Target/{1}{2}.pkl".format("Additional", tier, m))
        data = np.array(list(data2)+list(data3))
        targetVector = np.array(list(targetVector2)+list(targetVector3))
    # process data
    print("RNN_{} data.shape: {}".format(m, data.shape))
    # process target
    targetVector = np.array(targetVector)
    print("target.shape: {}".format(targetVector.shape))
```
